```python
# problem 1
import math
r = 10
print("Enter x, y coordinates")
x, y = map(int, input().split())
dist = math.sqrt(x*x + y*y)
if dist <= r:
    print("Dart hits on the board")
else:
    print("Dart does not hit the board")
```

```python
# Problem 5
# 1
a = int(input("Enter a number"))
o = input("Enter Operator")
b = int(input("Enter a number"))
if o == "+":
    print(a+b)
elif o == "-":
    print(a-b)
elif o == "*":
    print(a*b)
elif o == "/":
    print(a/b)
else:
    print("Enter valid operator")
```

```python
# 2
# swap using third variable
print("Enter two numbers")
a, b = map(str, input().split())
temp = a
a = b
b = temp
print(a, b)
```

```python
# 3
# swap without 3 rd variable
print("Enter two numbers")
a, b = map(str, input().split())
a, b = b, a
print(a, b)
```
